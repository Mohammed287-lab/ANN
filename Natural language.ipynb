```python
doc1 = "meeting ... management ... meeting ... management ... meeting "
doc1 += "... management ... meeting ... meeting"
query="management meeting fine"
query_length=len(query.split(" "))
vector=[0]*query_length
i=0
for qword in query.split(" "):
  for word in doc1.split(" "):
    if word==qword:
      vector[i]=vector[i]+1
  i=i+1
  print (vector)
```
