```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math


elements = { 'x':[1,2,3,4,5,6],'y':[6,7,3,5,4,2]}
df = pd.DataFrame(elements)

```

```python
x_mean = df['x'].mean()
y_mean = df['y'].mean()

df['x^2']= df['x']**2
df['y^2']= df['y']**2
df['xy']= df['x']*df['y']

```

```python
x_total = df['x^2'].sum()
y_total = df['y^2'].sum()
x_square_total = df['x^2'].sum()
y_square_total = df['y^2'].sum()
xy_total = df['xy'].sum()
```

```python
import math
n= len(df)
Slop= ( (n*xy_total)-  (x_total*y_total))  /((n*x_square_total)-(x_total**2))
Coefficient =(( y_total*x_square_total)-(x_total*xy_total))/((n*x_square_total)-(x_total**2))

df['y_predict'] = Coefficient + Slop * df['x']
df['Error'] = df['y'] - df['y_predict']

df ["squared error"] = df["Error"]**2
SSE = df["squared error"].sum()

df ["absolute error"] = df["Error"].abs()
SAE =df["absolute error"].sum()

RMSE = math.sqrt ((SSE/len(df["squared error"])))




```

```python
plt.scatter(df['x'],df['y'])
plt.plot(df['x'],df['y_predict'])

plt.show()
```

```python
x_input = input("Enter X :")
y_prediction = Coefficient + (Slop * float(x_input))
print('Predicted Y value:', y_prediction)
print('Sum of Squared Errors (SSE):', SSE)
print('Sum of Absolute Errors (SAE):', SAE)
print('Root Mean Squared Error (RMSE):', RMSE)
```
